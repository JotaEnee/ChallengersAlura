#1. An√°lise do faturamento


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Leitura e concatena√ß√£o dos dados
dfs = []
for i, url in enumerate(urls, 1):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()
    df['Loja'] = f'Loja {i}'
    dfs.append(df)

df_total = pd.concat(dfs, ignore_index=True)

# Convers√£o correta da data
df_total['Data da Compra'] = pd.to_datetime(df_total['Data da Compra'], dayfirst=True)

# C√°lculo do faturamento total por loja (soma da coluna 'Pre√ßo')
faturamento_loja = df_total.groupby('Loja')['Pre√ßo'].sum().reset_index()

print("üìà Faturamento Total por Loja:")
print(faturamento_loja)

# Gr√°fico de barras do faturamento por loja
plt.figure(figsize=(8,5))
plt.bar(faturamento_loja['Loja'], faturamento_loja['Pre√ßo'], color='green')
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (Soma dos Pre√ßos)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Lendo e juntando os dados
dfs = []
for i, url in enumerate(urls, 1):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()
    df['Loja'] = f'Loja {i}'
    dfs.append(df)

df_total = pd.concat(dfs, ignore_index=True)
df_total['Data da Compra'] = pd.to_datetime(df_total['Data da Compra'], dayfirst=True)

# Agrupamento do faturamento por loja e categoria (soma dos pre√ßos)
faturamento_categoria = df_total.groupby(['Loja', 'Categoria do Produto'])['Pre√ßo'].sum().reset_index()

print("\nüìä Faturamento por categoria e loja:")
print(faturamento_categoria)

# Gerar gr√°fico para cada loja
for loja in faturamento_categoria['Loja'].unique():
    dados_loja = faturamento_categoria[faturamento_categoria['Loja'] == loja]
    plt.figure(figsize=(10, 5))
    plt.bar(dados_loja['Categoria do Produto'], dados_loja['Pre√ßo'], color='skyblue')
    plt.title(f'Faturamento por Categoria - {loja}')
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Faturamento (Soma dos Pre√ßos)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# L√™ e prepara os dados
dfs = []
for i, url in enumerate(urls, start=1):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()
    df['Loja'] = f'Loja {i}'
    dfs.append(df)

df_total = pd.concat(dfs, ignore_index=True)

# Mostrar todas as colunas para confer√™ncia
print("Colunas dispon√≠veis no DataFrame:")
print(df_total.columns.tolist())

# Procurar colunas que contenham "avalia" (para detectar coluna de avalia√ß√£o)
avaliacao_cols = [col for col in df_total.columns if "avalia" in col.lower()]

if len(avaliacao_cols) == 0:
    print("\n‚ùå Nenhuma coluna relacionada a avalia√ß√£o foi encontrada.")
else:
    col_avaliacao = avaliacao_cols[0]
    print(f"\n‚úÖ Coluna de avalia√ß√£o detectada: '{col_avaliacao}'")

    # Calcular m√©dia de avalia√ß√£o por loja
    media_avaliacao = df_total.groupby('Loja')[col_avaliacao].mean().reset_index()
    media_avaliacao = media_avaliacao.sort_values(by=col_avaliacao, ascending=False)

    print("\n‚≠ê M√©dia de Avalia√ß√£o por Loja:")
    print(media_avaliacao)

    # Plotar gr√°fico
    plt.figure(figsize=(8,5))
    plt.bar(media_avaliacao['Loja'], media_avaliacao[col_avaliacao], color='orange')
    plt.ylim(0, 5)
    plt.title('M√©dia de Avalia√ß√£o das Lojas')
    plt.ylabel('Avalia√ß√£o (0 a 5)')
    plt.xlabel('Loja')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

# Lendo e limpando dados
dataframes = []
for i, url in enumerate(urls, 1):
    df = pd.read_csv(url)
    df.columns = df.columns.str.strip()  # limpa espa√ßos em branco nos nomes das colunas
    df['Loja'] = f'Loja {i}'
    dataframes.append(df)

df_total = pd.concat(dataframes, ignore_index=True)

# Mostrar as colunas para voc√™ conferir os nomes
print("Colunas do DataFrame combinado:")
print(df_total.columns.tolist())

# Detectar automaticamente a coluna de quantidade
col_qtd = None
for col in df_total.columns:
    if 'quantidade' in col.lower():
        col_qtd = col
        break

if col_qtd is None:
    raise ValueError("Coluna de quantidade n√£o encontrada no DataFrame. Verifique os nomes das colunas acima.")

# Detectar coluna de avalia√ß√£o da loja, se existir
col_avaliacao = None
for col in df_total.columns:
    if 'avalia' in col.lower():  # pega avalia√ß√£o da loja ou similares
        col_avaliacao = col
        break

# Mostrar m√©dia de avalia√ß√£o se coluna existir
if col_avaliacao:
    media_avaliacao = df_total.groupby('Loja')[col_avaliacao].mean().reset_index()
    media_avaliacao = media_avaliacao.sort_values(by=col_avaliacao, ascending=False)

    print(f"\n‚≠ê M√©dia de Avalia√ß√£o por Loja (coluna: '{col_avaliacao}'):")
    print(media_avaliacao)

    plt.figure(figsize=(8, 5))
    plt.bar(media_avaliacao['Loja'], media_avaliacao[col_avaliacao], color='orange')
    plt.ylim(0, 5)
    plt.title('M√©dia de Avalia√ß√£o das Lojas')
    plt.ylabel('Avalia√ß√£o (0 a 5)')
    plt.xlabel('Loja')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
else:
    print("\n‚ùå Coluna de avalia√ß√£o da loja n√£o encontrada. Pulando essa an√°lise.")

# Fun√ß√£o para produtos mais e menos vendidos por loja
def produtos_mais_menos_vendidos(df, loja_nome, coluna_quantidade):
    dados_loja = df[df['Loja'] == loja_nome]
    resumo = dados_loja.groupby('Produto')[coluna_quantidade].sum().reset_index()
    resumo = resumo.sort_values(by=coluna_quantidade, ascending=False)

    print(f"\nüìà Top 5 produtos mais vendidos - {loja_nome}:")
    print(resumo.head(5).to_string(index=False))

    print(f"\nüìâ Top 5 produtos menos vendidos - {loja_nome}:")
    print(resumo.tail(5).to_string(index=False))

    # Gr√°fico Top 5 produtos mais vendidos
    plt.figure(figsize=(8, 4))
    plt.bar(resumo.head(5)['Produto'], resumo.head(5)[coluna_quantidade], color='skyblue')
    plt.title(f'Top 5 Produtos Mais Vendidos - {loja_nome}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Gr√°fico Top 5 produtos menos vendidos
    plt.figure(figsize=(8, 4))
    plt.bar(resumo.tail(5)['Produto'], resumo.tail(5)[coluna_quantidade], color='salmon')
    plt.title(f'Top 5 Produtos Menos Vendidos - {loja_nome}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Aplicar para cada loja
for loja in df_total['Loja'].unique():
    produtos_mais_menos_vendidos(df_total, loja, col_qtd)
```

# 5. Frete M√©dio por Loja


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Ler os dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Limpar nomes das colunas
for df in [loja1, loja2, loja3, loja4]:
    df.columns = df.columns.str.strip()

# Identificar lojas
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Unir dados
df_total = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Calcular frete m√©dio por loja
frete_medio = df_total.groupby('Loja')['Frete'].mean().reset_index()

print("\nüöö Frete m√©dio por loja (R$):")
print(frete_medio)

# Gr√°fico de barras
plt.figure(figsize=(8, 5))
plt.bar(frete_medio['Loja'], frete_medio['Frete'], color='green')
plt.title('Frete M√©dio por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete M√©dio (R$)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
